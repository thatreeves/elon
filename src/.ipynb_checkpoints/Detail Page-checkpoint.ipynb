```python
# importing the required modules
import os
import glob

import csv

selfclass="from-me"
otherclass="from-them"

frame = '''
<!DOCTYPE html>
<html lang="en">
  <head>
  <!-- Primary Meta Tags -->
<title>Inside Elon Musk's Messages</title>
<meta name="title" content="Inside Elon Musk's Messages">
<meta name="description" content="Inside Elon's leaked iMessages from his Twitter lawsuit.">

<!-- Open Graph / Facebook -->
<meta property="og:type" content="website">
<meta property="og:url" content="http://muskmessages.com/">
<meta property="og:title" content="Inside Elon Musk's Messages">
<meta property="og:description" content="Inside Elon's leaked iMessages from his Twitter lawsuit.">
<meta property="og:image" content="">

<!-- Twitter -->
<meta property="twitter:card" content="summary_large_image">
<meta property="twitter:url" content="http://muskmessages.com/">
<meta property="twitter:title" content="Inside Elon Musk's Messages">
<meta property="twitter:description" content="Inside Elon's leaked iMessages from his Twitter lawsuit.">
<meta property="twitter:image" content="">

    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <meta http-equiv="X-UA-Compatible" content="ie=edge">
    <title>Musk Message</title>
    <link rel="stylesheet" href="../dist/output.css">
    <link href="https://fonts.cdnfonts.com/css/sf-pro-display" rel="stylesheet">
    
    
<!-- Google tag (gtag.js) -->
<script async src="https://www.googletagmanager.com/gtag/js?id=G-V2YWT8DJ14"></script>
<script>
  window.dataLayer = window.dataLayer || [];
  function gtag(){dataLayer.push(arguments);}
  gtag('js', new Date());

  gtag('config', 'G-V2YWT8DJ14');
</script>    
    
    
  </head>
  <body class="bg-gray-800">
  <nav class="flex flex-col justify-center w-full text-white bg-gray-900">
        <a href="/">
            <h1 class="text-lg font-bold text-center my-2">Elon's iMessages</h1>
        </a>
        <p class="mb-10 text-sm text-center md:hidden">We've aimed to be as accurate as possible, but we can't guarantee accuracy or validity. Also, some of the documents were redacted, so some messages don't make sense. </p>
    </nav>
    <div class="w-full max-w-screen-lg mx-auto flex flex-row mt-4">

    <div class="flex w-full min-h-[90vh] max-w-[600px]  lg:w-1/2">
            <div class="flex w-full max-w-[600px] ">
                
                <div class="flex imessage w-full pt-0">
                    <div class="w-full flex flex-col sticky top-0 z-50 bg-blend-screen backdrop-blur-sm bg-white/90 pt-4 pb-2 text-base border-b">
                        <div class="w-full flex flex-col self-center">
                            <div class="flex font-bold self-center text-base p-0 m-0 font-['SF Pro Display']">{1}</div>
                        </div>
                        <div class="w-full flex flex-row justify-between">
                            <div class="flex max-w-[40px]"><a href="./../"><img  src="../img/Arrow.png" /></a></div>
                            <div class="flex max-w-[30px] flex-col justify-center"><div><img  src="../img/Facetime.png" /></a></div></div>
                        </div>
                    </div>

                    <div class="flex w-full flex-col">
                        {0}
                    </div>
                    
                    </div>
  
              </div>

        </div>

        
        <div class="hidden lg:flex w-1/2 text-white">
            <div class="flex flex-col w-full p-3 m-6 bg-grey-100">
                <div class="flex flex-col ">
                    <h1 class="text-lg font-bold my-2">Elon's iMessages</h1>
                    <p class="mb-10">The messages are taken from public records and filings from Elon Musk's upcoming court case with Twitter.</p>
                    <p class="mb-10 text-sm">We've aimed to be as accurate as possible, but we can't guarantee accuracy or validity. The documents were redacted, so some messages don't make sense. </p>
                </div>
                
                <div class="text-md font-bold">Highlights</div>
                    <a href="../d/36.html" class="p-4 hover:bg-gray-900 border-gray-300 rounded-md my-4 border">
                        <h3 class="font-bold">ðŸŽ‰ Joe Rogan Messages About a Party ðŸŽ‰</h3>
                        <p class="text-sm">Read</p>
                    <a>
                    <a href="../d/18.html" class="p-4 hover:bg-gray-900 border-gray-300 rounded-md my-4 border">
                        <h3 class="font-bold">Goofin with Gayle King</h3>
                        <p class="text-sm">Read</p>
                    <a>
                    <a href="../d/34.html" class="p-4 hover:bg-gray-900 border-gray-300 rounded-md my-4 border">
                        <h3 class="font-bold">Jason "Magic Thumbs" Calacanis</h3>
                        <p class="text-base">Who send 65+ messages</p>
                        <p class="text-sm">Read</p>
                    <a>
                    <a href="../d/20.html" class="p-4 hover:bg-gray-900 border-gray-300 rounded-md my-4 border">
                        <h3 class="font-bold">Nice chats with Jack Jack (twitter founder guy)</h3>
                        <p class="text-base"></p>
                        <p class="text-sm">Read</p>
                    <a>
                    <a href="mailto:matthew@muskmessages">matthew@muskmessages.com</a>
                </div>
            </div>
        </div>

    </div>

    
  </body>
</html>
'''

messageFrame = '''
<div class="text-xs text-grey-600 mt-2">{3}{2}</div>
<p class="{0}">{1}</p>
'''
# final = frame.format(messageFrame)



cwd =""
cwd = os.getcwd()

print(cwd)
path = cwd
os.chdir(path)
all_files = glob.glob('*.csv')

for file in all_files:
    html ="../../d/"
    filename=""
    filename = html+os.path.splitext(file)[0]+".html"
    

    with open(file, newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        messageBlock = ""
        for row in reader:
            recipient = row['ParticipantsClean']
            sender = ""
            if row['Group'] == "Yes" and row['Sender'] != "Self":
                sender ='<span>'+row['Sender']+' - </span>'
            if row['Sender'] == "Self":
                messageBlock = messageBlock+(messageFrame.format(selfclass, row['Content'],row['Date'],sender))
            else:
                messageBlock = messageBlock+(messageFrame.format(otherclass, row['Content'],row['Date'],sender))
            
            finalHTML = frame.format(messageBlock, recipient)
    
    
    # Creating an HTML file
    Func = open(filename,"w")

    # Adding input data to the HTML file
    Func.write(finalHTML)

    # Saving the data into the HTML file
    Func.close()
```
